```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
neural_data = pd.read_csv(r"D:\Engineering\Sem 7\Machine Learning\Take home assignmet\Nuural_data.csv")
```

```python
neural_data.head
```

```python
# check shape of the data set
neural_data.shape
```

```python
# change data frame labels
# index no is odd number group class 1 and class 3 together
# class 1 and class 3 = class_A, class 2 and class 4 = class_B

neural_data['AB'].replace(1, 'class_A',inplace=True)
neural_data['AB'].replace(3, 'class_A',inplace=True)
neural_data['AB'].replace(2, 'class_B',inplace=True)
neural_data['AB'].replace(4, 'class_B',inplace=True)
```

```python
neural_data["AB"]
```

```python
# drop the labels
X = neural_data.drop("AB",axis =1) 
```

```python
X.tail
```

```python
# with out applying data preprocessing techniques
# sklearn pca 
from sklearn.decomposition import PCA
pca_neural = PCA(n_components=2)
principalComponents_neural = pca_neural.fit_transform(X)
```

```python
# convert to data frame
principal_neural_Df = pd.DataFrame(data = principalComponents_neural
             , columns = ['principal component 1', 'principal component 2'])
```

```python
principal_neural_Df
```

```python
# draw mean value of each coloumn
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component - 1',fontsize=20)
plt.ylabel('Principal Component - 2',fontsize=20)
plt.title("Principal Component Analysis of Breast Cancer Dataset",fontsize=20)
targets = ['class_A', 'class_B']
colors = ['r', 'g',]
for target, color in zip(targets,colors):
    indicesToKeep = neural_data['AB'] == target
    plt.scatter(principal_neural_Df.loc[indicesToKeep, 'principal component 1']
               , principal_neural_Df.loc[indicesToKeep, 'principal component 2'], c = color, s = 50)

plt.legend(targets,prop={'size': 15})
```
