### Manipluating tensors
** Basic Operation ** 

`+`,`-`,`*`,`/`

```python
tensor = tf.constant(
    [[10,7],
    [8,6]]
)

tensor + 10  # original tensor vales are not changed
```

```python
tensor # original tensor remain same
# can use tensorflow built in function too
tf.multiply(tensor,10)
```

** Matrix Multiplication **

```python
print(tensor)
```

```python
tf.matmul(tensor,tensor)
```

```python
x = tf.constant(
    [
        [1,2],
        [3,4],
        [6,8]
    ]
)

y = tf.constant(
    [
        [4,8],
        [2,6],
        [9,19]
    ]
)

x,y
```

```python
# matrix multiplycation using python operator
# first we have to reshape the matrix

x @ tf.reshape(y, shape=(2,3))
```

```python
# using built in fuction
tf.matmul(x, tf.reshape(y, shape=(2,3)))
```

```python
# transpose of a matrix
tf.matmul(x, tf.transpose(y))
```
